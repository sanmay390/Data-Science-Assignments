```python
import pandas as pd
from mftool import Mftool
result = Mftool().get_available_schemes('ICICI Prudential Large')
print(result)
```

```python
import requests as rq
import json as js
from pandas import json_normalize
```

```python
hdfc_response = rq.get('https://api.mfapi.in/mf/105758', allow_redirects=True)
axis_response = rq.get('https://api.mfapi.in/mf/114564', allow_redirects=True)
icici_response = rq.get('https://api.mfapi.in/mf/100349', allow_redirects=True)
print(hdfc_response.headers.get('content-type'))
```

```python
hdfc_content = hdfc_response.content
axis_content = axis_response.content
icici_content = icici_response.content
print(hdfc_content)
print(axis_content)
print(icici_content)
```

```python
hdfc_data = js.loads(hdfc_content)
axis_data = js.loads(axis_content)
icici_data = js.loads(icici_content)
hdfc_data = json_normalize(hdfc_data)
axis_data = json_normalize(axis_data)
icici_data = json_normalize(icici_data)
```

```python
hdfc_df = pd.DataFrame(hdfc_data['data'][0])
axis_df = pd.DataFrame(axis_data['data'][0])
icici_df = pd.DataFrame(icici_data['data'][0])
```

```python
hdfc_df.head(30)
```

```python
hdfc_req_df = hdfc_df[6:25]
hdfc_req_df
```

```python
axis_req_df = axis_df[6:25]
axis_req_df
```

```python
icici_req_df = icici_df[6:25]
icici_req_df
```

```python
hdfc_present_nav = float(hdfc_req_df['nav'][24])
hdfc_initial_nav = float(hdfc_req_df['nav'][6])
axis_present_nav = float(axis_req_df['nav'][24])
axis_initial_nav = float(axis_req_df['nav'][6])
icici_present_nav = float(icici_req_df['nav'][24])
icici_initial_nav = float(icici_req_df['nav'][6])
```

```python
hdfc_abs_ret = ((hdfc_present_nav - hdfc_initial_nav) / hdfc_initial_nav) * 100
axis_abs_ret = ((axis_present_nav - axis_initial_nav) / axis_initial_nav) * 100
icici_abs_ret = ((icici_present_nav - icici_initial_nav) / icici_initial_nav) * 100
print("HDFC Mid-Cap Opportunities Fund - Growth Plan Absolute return:",hdfc_abs_ret)
print("Axis Midcap Fund - Regular Plan - Growth Absolute return:",axis_abs_ret)
print("ICICI Prudential Large & Mid Cap Fund - Growth Absolute return:",icici_abs_ret)
```

HDFC Mid-Cap Opportunities Fund - Growth Plan performed well in March as compared to other 2 mutual funds.
